numpy是Python专门处理高维数组(high dimensional array)的计算的包

```python
my_array = np.arange(100000)
my_list = list(range(100000))
# my_array,my_list
```

```python
%time for _ in range(10): my_array2 = my_array *  2
```

```python
%time for _ in range(10): my_list2 = [x * 2 for x in my_list]
```
