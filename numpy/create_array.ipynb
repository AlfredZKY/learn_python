```python
import numpy as np
```

数组 (array) 是相同类型的元素 (element) 的集合所组成数据结构 (data structure)  
注意一个关键字 axis，中文叫「轴」，一个数组是多少维度就有多少根轴。由于 Python 计数都是从 0 开始的.

数组创建的几种方法:  
1.按步就班的np.array()用于列表和元祖上  
2.定隔定点的np.arange()和np.linespace()  
3.一步登天的np.ones(),np.zeros()和np.randon.random()

```python
# 按步就班法
# 注意，numpy 数组的输出都带有 array() 的字样，里面的元素用「中括号 []」框住。
l = [3.45,5,2,8,4.2]
np.array(l)
```

定隔定点的np.arange()和np.linespace()
更常见的两种创建
- 定隔的arange:固定元素大小间隔
- 定点的linespace:固定元素个数
